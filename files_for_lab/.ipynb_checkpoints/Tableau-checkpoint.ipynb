```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
df.head()
```

```python
sns.countplot(data = df,x='Gender')
plt.title('number of customers per Gender')
plt.xlabel('Gender')
plt.ylabel('No.of Customers')
plt.show()
```

```python
sns.countplot(data = df,x='EmploymentStatus',hue = 'Gender')
plt.title('Number of Customers by EmploymentStatus and Gender')
plt.xlabel('EmploymentStatus')
plt.ylabel('No.of Customers')
plt.legend(title = 'Gender')
plt.xticks(rotation=45)  
plt.tight_layout()  
plt.show()
```
